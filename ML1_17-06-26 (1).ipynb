```python
import numpy as np
arr=np.arange(1,21)
print("Array:")
print(arr)
print("\nShape:",arr.shape)
print("Maximum:",arr.max())
print("Mininum:",arr.min())
print("Mean:",arr.mean())
print("Standard Deviation:",arr.std())
matrix=arr.reshape(4,5)
print("\n4x5 Matrix:")
print(matrix)
print("\nSecond Row:")
print(matrix[1])
print("\nThied Column:")
print(matrix[:,2])
```

```python
import pandas as pd
data={
    "Roll No":[101,102,103,104,105],
    "Name":["Ravi","Priya","Kiran","Anu","Sita"],
    "Marks":[85,92,78,88,95],
}
df=pd.DataFrame(data)
print("DataFrame:")
print(df)
print("\nFirst Three Records:")
print(df.head(3))
print("\nAverage Marks:",df["Marks"].mean())
print("\nHighest Marks Student:")
print(df.loc[df["Marks"].idxmax()])
print("\nLowest Marks Student:")
print(df.loc[df["Marks"].idxmin()])
def grade(mark):
    if mark>=90:
        return "A"
    elif mark>=80:
        return "B"
    else:
        return "C"
df["Grade"]=df["Marks"].apply(grade)
print("\nDataFrame with Grade:")
print(df)
```

```python
import pandas as pd
data={
    "Product":["Laptop","Mouse","Keyboerd","Monitor"],
    "Price":[50000,500,1200,10000],
    "Quantity":[10,50,30,15],
}
df=pd.DataFrame(data)
df["Total Value"]=df["Price"]*df["Quantity"]
print(df)
print("\nHighest Total Value Product:")
print(df.loc[df["Total Value"].idxmax()])
print("\nOverall Inventory Value:")
print(df["Total Value"].sum())
```

```python
import numpy as np
from sklearn.datasets import load_iris
iris=load_iris()
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df["Species"]=pd.Categorical.from_codes(iris.target,iris.target_names)
print("DataFrame:")
print(df)
print("\nFirst Five Rows:")
print(df.head())
print("\nRows and Columns:")
print(df.shape)
print("\nColumn Names:")
print(df.columns)
print("\nMissing Values:")
print(df.isnull().sum())
print("\nSummary Statistics:")
print(df.describe())
print("\nSamples in Each Class:")
print(df["Species"].value_counts())
```
